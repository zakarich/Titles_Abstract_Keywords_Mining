```python
import os
import shutil

import re
import time
import pandas as pd

IEEE = pd.read_excel('IEEE.xlsx')
SCOPUS = pd.read_excel('SCOPUS.xlsx')
SPRINGER = pd.read_excel('SPRINGER.xlsx')
```

```python
IEEEtrans = IEEE.transform(lambda x: x.fillna('XXX'))
SCOPUStrans = SCOPUS.transform(lambda x: x.fillna('XXX'))
SPRINGERtrans = SPRINGER.transform(lambda x: x.fillna('XXX'))

IEEEtrans.drop(IEEEtrans.index[IEEEtrans['Publication_Year'] == 'XXX'], inplace = True)
SCOPUStrans.drop(SCOPUStrans.index[SCOPUStrans['Publication_Year'] == 'XXX'], inplace = True)
SPRINGERtrans.drop(SPRINGERtrans.index[SPRINGERtrans['Publication_Year'] == 'XXX'], inplace = True)

IEEEtrans.drop(IEEEtrans.index[IEEEtrans['Publication_Year'] == 2021.0], inplace = True)
SCOPUStrans.drop(SCOPUStrans.index[SCOPUStrans['Publication_Year'] == 2021.0], inplace = True)
SPRINGERtrans.drop(SPRINGERtrans.index[SPRINGERtrans['Publication_Year'] == 2021.0], inplace = True)

IEEEtrans['Publication_Year'] = IEEEtrans['Publication_Year'].astype(int)
SCOPUStrans['Publication_Year'] = SCOPUStrans['Publication_Year'].astype(int)
SPRINGERtrans['Publication_Year'] = SPRINGERtrans['Publication_Year'].astype(int)


groupedIEEE = IEEEtrans.groupby('Publication_Year')
gr_IE = groupedIEEE.count()

groupedSCOPUS = SCOPUStrans.groupby('Publication_Year')
gr_SC = groupedSCOPUS.count()

groupedSPRINGER = SPRINGERtrans.groupby('Publication_Year')
gr_SP = groupedSPRINGER.count()

df1 = pd.DataFrame()
df2 = pd.DataFrame()
df3 = pd.DataFrame()

df1['YEAR'] = gr_IE.index.values
df1['IEEE'] = gr_IE['Title'].values

df2['YEAR'] = gr_SC.index.values
df2['SCOPUS'] = gr_SC['Title'].values

df3['YEAR'] = gr_SP.index.values
df3['SPRINGER'] = gr_SP['Title'].values

df1
```

```python
import matplotlib.pyplot as plt
#Get current axis 
ax = plt.gca()  
  
# line plot for math marks 
df1.plot(kind = 'line', 
        x = 'YEAR', 
        y = 'IEEE', 
        color = 'green',ax = ax) 
  
# line plot for physics marks 
df2.plot(
    kind = 'line',
    x = 'YEAR', 
    y = 'SCOPUS', 
    color = 'blue',ax = ax) 
  
# line plot for chemistry marks 
df3.plot(kind = 'line',
           x = 'YEAR', 
           y = 'SPRINGER', 
           color = 'black',ax = ax) 
  
# set the title 
plt.title('research evolution over the years') 
  
# show the plot 
plt.show() 
```
