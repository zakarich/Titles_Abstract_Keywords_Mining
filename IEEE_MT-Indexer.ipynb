```python
import os
import shutil

import re
import time
import pandas as pd

#encoding="utf8"

csv_fileIEEE = pd.read_csv('19-07-21-Scopus/IEEE-Failed.csv', header=0, engine='python', warn_bad_lines=True, error_bad_lines=False)


csv_fileIEEE
```

```python
csv_fileIEEE.dropna(subset = ["Author Keywords"], inplace=True)
csv_fileIEEE.dropna(subset = ["Document Title"], inplace=True)
csv_fileIEEE.reset_index(drop=True, inplace=True)
csv_fileIEEE
```

```python
csv_fileIEEE['Author Keywords'][1]
```

```python
import string
csv_fileIEEE['CleanTitles'] = csv_fileIEEE['Document Title'].str.lower()
csv_fileIEEE['CleanAuthKeys'] = csv_fileIEEE['Author Keywords'].str.lower()
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = str(text).replace(punctuation, ' ')
    return text

csv_fileIEEE['CleanTitles'] = csv_fileIEEE['CleanTitles'].apply(remove_punctuations)
csv_fileIEEE['CleanAuthKeys'] = csv_fileIEEE['CleanAuthKeys'].apply(remove_punctuations)

csv_fileIEEE['CleanAuthKeys'][1]
```

```python
PatternsInn = []
#Remove All words that could mess our classification
Words1 = ['machine','automatic']
Words2 = ['translation','translator','translators','translating','translate','translated']
#Lang = [word.lower() for word in Languages]
# Recognition for IEEE
for i in range(len(csv_fileIEEE["CleanTitles"])): 
    #res = re.sub(r'[^\w\s]', ' ', df.astype(str)["Document Title"][i]+df.astype(str)["Abstract"][i]+df.astype(str)["Author Keywords"][i])
    #res = re.sub(r'[^\w\s]', ' ', str(csv_fileIEEE["Titles & Keywords"][i]))
    #texto = res.lower().split()
    decision = 0 
    for j in range(len(Words2)):
        a = re.sub(r'[^\w\s]', ' ', str(csv_fileIEEE["CleanTitles"][i]))
        b1 = r'\b' + r'\b.*\b'.join(re.escape(word) for word in Words1[0].split(' ')) + r'\b'
        b2 = r'\b' + r'\b.*\b'.join(re.escape(word) for word in Words1[1].split(' ')) + r'\b'
        b = r'\b' + r'\b.*\b'.join(re.escape(word) for word in Words2[j].split(' ')) + r'\b'
        c = bool((re.search(b,a) and re.search(b1,a)) or (re.search(b,a) and re.search(b2,a)) or (re.search('mt',a)) or (re.search('smt',a)) or (re.search('nmt',a)) or (re.search('rbmt',a)) or (re.search('hmt',a)))
        if c == True:
            decision = 1
            break
    PatternsInn.append(decision)
    
csv_fileIEEE["TitlesPatterns"] = PatternsInn
print("The number of Patterns that the program can recognize :"+str(len(Words2) * len(Words1)))
csv_fileIEEE
```
